```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#from pandas_datareader import eurostat
import seaborn as sns
import eurostat
```

```python
life_exp = eurostat.get_data_df('sdg_03_10')
smoking  = eurostat.get_data_df('sdg_03_30')
obesity =  eurostat.get_data_df('sdg_02_10')
gdppc = eurostat.get_data_df('sdg_08_10')
life_exp = life_exp.melt(id_vars = ['unit','sex','age', 'geo\\time'], var_name = 'Year', value_name = 'Life Expectancy')
life_exp = life_exp.pivot(index = ['geo\\time','Year'], columns = 'sex', values ='Life Expectancy')
life_exp.columns = ['Life Expectancy Females', 'Life Expectancy Males', 'Life Expectancy All']
smoking = smoking.melt(id_vars = ['unit','sex','age', 'geo\\time'], var_name = 'Year', value_name = 'Smoking')
smoking = smoking.pivot(index = ['geo\\time','Year'], columns = 'sex', values ='Smoking')
smoking.columns = ['Smoking Females', 'Smoking Males', 'Smoking All']
obesity = obesity.melt(id_vars = ['unit','bmi', 'geo\\time'], var_name = 'Year', value_name = 'Obesity')
obesity = obesity.pivot(index = ['geo\\time','Year'], columns = 'bmi', values ='Obesity')
obesity.columns  = ['Pre-obese','Overweight', 'Obese']
gdppc = gdppc.melt(id_vars = ['unit','na_item', 'geo\\time'], var_name = 'Year', value_name = 'GDP per capita')
gdppc = gdppc.pivot(index = ['geo\\time','Year'], columns = 'unit', values ='GDP per capita')
gdppc.columns = ['GDP per capita in 2010 EUR', 'GDP per capita growth']
```

```python
life_exp = life_exp.groupby(by='geo\\time').mean()
smoking = smoking.groupby(by='geo\\time').mean()
obesity = obesity.groupby(by='geo\\time').mean()
gdppc = gdppc.groupby(by = 'geo\\time').mean()
```

```python
a = pd.merge(left = life_exp, right = smoking, how = 'inner', left_index = True, right_index = True)
b = pd.merge(left = a, right = obesity, how = 'inner', left_index = True, right_index = True)
data = pd.merge(left = b, right = gdppc, how = 'inner', left_index = True, right_index = True)
data = data.reset_index()
names = data.columns.to_list()
names[0] = 'Country'
data.columns = names
```
